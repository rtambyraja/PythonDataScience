In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap


#Note the version of matplotlib imported is 2.0.0


np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
```

```python
%matplotlib notebook
#apparently this above command is what creates an interactive plot? Not sure about this.
```

```python


#Basic x and y data for the plot
x_vals = [1992, 1993, 1994, 1995]
bar_vals2 = [df.loc[j].mean() for j in df.index]

#Calculating 95% Confidence Interval (CI) - Need to calculate standard error of the mean, then multiply by 1.96 (Z-score). 
#The CI is each mean +/- the standard error of the mean * Z-score
stand_err_mean = [(df.loc[j].std()/np.sqrt(3650)) for j in df.index]
Zscore = [j* 1.96 for j in stand_err_mean]

#More plot parameters, like color scheme - RdBu called out in docs as a diverging Blue-Red scheme, that's what we want.
cmap = cm.get_cmap('RdBu')

#Trying to create a Normalized color map so I can change the color based on a calculated value
norm = mpl.colors.Normalize(vmin=-1, vmax=1)
```

```python



#I think the plt.ion() command should turn on interactive, per the documentation
plt.ion()

#Setting my plot as a variable so I can call individual columns (years) more easily
#Note that for yerr, I only specified one value, as yerr does the +/- part for me
bar_plot = plt.bar(x_vals, bar_vals2, tick_label= x_vals, yerr=Zscore, capsize=7)



#Now trying to create the functions that will define interactivity. I want a function that prints a horizontal line on the plot  
# at the mouseclick's y-value (Y). It should then shade each bar according to the cmap. Full red if Y is below the CI of the bar,
# and full blue if Y is above the CI. Shaded gradient if Y is within the CI. 

def onclick(event):
    for n in range(4):
        barcolor = cmap(norm(bar_vals2[n] - event.ydata))
        print(barcolor)
        bar_plot[n].set_color(barcolor)
    plt.axhline(y=event.ydata, ls='--')

    
#I think this command is supposed to connect the figure to the onclick function:
# tell mpl_connect we want to pass a 'button_press_event' into onclick when the event is detected
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
plt.show()
```
